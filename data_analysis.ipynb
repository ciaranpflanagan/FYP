```python
import pandas as pd
import numpy as np
```

```python
df_2021 = pd.read_csv('data/2021.csv')
df_2021.dropna(inplace=True)
df_2021.tail(20)
```

```python
df_2021.describe()
```

```python
d = {'10cm' : df_2021['BD Moisture 0-10cm'],
     '20cm' : df_2021['BD Moisture 10-20 cm'],
     '30cm' : df_2021['BD Moisture 20-30 cm']
    }

df = pd.DataFrame(d)

df.plot()
```

Results above show what we would expect, first layer of the soil contains less moisture then the second two layers. However it also shows that there isn't much of a difference between the second and third layers.

```python
d = {'Emergence' : df_2021['Emergence'],
     'Yield' : df_2021['Combine Yield (t/ac)']
    }

df = pd.DataFrame(d)

df.plot()
```

```python
correlation = df_2021['Emergence'].corr(df_2021['Combine Yield (t/ac)'])
correlation
```

# Moisture & Emergence / Yield

```python
average_moisture = df_2021[['BD Moisture 0-10cm', 'BD Moisture 10-20 cm', 'BD Moisture 20-30 cm']].mean(axis=1)

d = {'Emergence' : df_2021['Emergence'],
     'Yield' : df_2021['Combine Yield (t/ac)'],
     'Avg Moisture' : average_moisture
    }

df = pd.DataFrame(d)

df.plot()
```
