```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# Methods

```python
# Plots data
def plotYearData(df, title):
    df = pd.DataFrame(df)
    
    plt.plot(df) 
    plt.xlabel('Plot Number') 
    plt.ylabel('Soil Moisture Content %') 
    
    # displaying the title
    plt.title(title)
    plt.legend(['10cm', '20cm', '30cm'])
    
    plt.show()
```

# Reading Data


```python
df_2021 = pd.read_csv('data/2021.csv')
df_2021.dropna(inplace=True)
df_2021.head()

df_2020 = pd.read_csv('data/2020.csv')
df_2020.dropna(inplace=True)
df_2020.head()
```

```python
df_2021.describe()
```

```python
df_2020.describe()
```

```python
d = {
    '10cm' : df_2021['BD Moisture 0-10cm'],
    '20cm' : df_2021['BD Moisture 10-20 cm'],
    '30cm' : df_2021['BD Moisture 20-30 cm']
}

plotYearData(d, 'Soil Moisture Level 2021')
```

```python
moisture_2020 = df_2020[['BD Moisture 0-10cm', 'BD Moisture 10-20 cm', 'BD Moisture 20-30 cm']]
cleaned_moisture_2020 = moisture_2020.loc[(moisture_2020 != 0).any(axis = 1)]

d = {
    '10cm' : cleaned_moisture_2020['BD Moisture 0-10cm'],
    '20cm' : cleaned_moisture_2020['BD Moisture 10-20 cm'],
    '30cm' : cleaned_moisture_2020['BD Moisture 20-30 cm'],
}

plotYearData(d, 'Soil Moisture Level 2020')
```

# Moisture & Emergence / Yield

```python
average_moisture = df_2021[['BD Moisture 0-10cm', 'BD Moisture 10-20 cm', 'BD Moisture 20-30 cm']].mean(axis=1)

d = {'Emergence' : df_2021['Emergence'],
     'Yield' : df_2021['Combine Yield (t/ac)'],
     'Avg Moisture' : average_moisture
    }

df = pd.DataFrame(d)

df.plot()
```
