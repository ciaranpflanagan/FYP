```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# Methods

```python
# Plots data
def plotYearData(df, title):
    df = pd.DataFrame(df)
    
    plt.plot(df) 
    plt.xlabel('Plot Number') 
    plt.ylabel('Soil Moisture Content %') 
    
    # displaying the title
    plt.title(title)
    plt.legend(['10cm', '20cm', '30cm'])
    
    plt.show()
```

# Reading Data


```python
df_2021 = pd.read_csv('data/2021.csv')
df_2021.dropna(inplace=True)
df_2021.head()

df_2020 = pd.read_csv('data/2020.csv')
df_2020.dropna(inplace=True)
df_2020.head()
```

```python
d = {
    '10cm' : df_2021['BD Moisture 0-10cm'],
    '20cm' : df_2021['BD Moisture 10-20 cm'],
    '30cm' : df_2021['BD Moisture 20-30 cm']
}

plotYearData(d, 'Soil Moisture Level 2021')
```

```python
moisture_2020 = df_2020[['BD Moisture 0-10cm', 'BD Moisture 10-20 cm', 'BD Moisture 20-30 cm']]
cleaned_moisture_2020 = moisture_2020.loc[(moisture_2020 != 0).any(axis = 1)]

d = {
    '10cm' : cleaned_moisture_2020['BD Moisture 0-10cm'],
    '20cm' : cleaned_moisture_2020['BD Moisture 10-20 cm'],
    '30cm' : cleaned_moisture_2020['BD Moisture 20-30 cm'],
}

plotYearData(d, 'Soil Moisture Level 2020')
```

# Treatments
Each plot is assigned a treatment type. Each treatment is a combination of three attributes, operation type, pressure level and moisture level.

**Operation Type:** Till, Plough    
**Pressure Level:** High, Low    
**Moisture Level:** High, Low    

```python
df_2021['Treatment']
```

```python
# 2021
ploughed_2021 = df_2021.loc[df_2021['Treatment'].str.startswith('PH', na=False)]
ploughed_2021 = ploughed_2021.append(df_2021.loc[df_2021['Treatment'].str.startswith('PL', na=False)])

tilled_2021 = df_2021.loc[df_2021['Treatment'].str.startswith('MH', na=False)]
tilled_2021 = tilled_2021.append(df_2021.loc[df_2021['Treatment'].str.startswith('ML', na=False)])

# 2020
ploughed_2020 = df_2020.loc[df_2020['Treatment'].str.startswith('PH', na=False)]
ploughed_2020 = ploughed_2020.append(df_2020.loc[df_2020['Treatment'].str.startswith('PL', na=False)])

tilled_2020 = df_2020.loc[df_2020['Treatment'].str.startswith('MH', na=False)]
tilled_2020 = tilled_2020.append(df_2020.loc[df_2020['Treatment'].str.startswith('ML', na=False)])
```

```python
tilled_2020.head()
```

## Plotting Average Yields

```python
# Ploughed Data
targets_2020 = ploughed_2020[['Emergence', 'Combine Yield (t/ac)']]
targets_2020 = targets_2020.loc[(targets_2020 != 0).all(axis = 1)]
targets_2021 = ploughed_2021[['Emergence', 'Combine Yield (t/ac)']]
targets_2021 = targets_2021.loc[(targets_2021 != 0).all(axis = 1)]

d = {'Emergence \'20' : targets_2020['Emergence'],
     'Yield \'20' : targets_2020['Combine Yield (t/ac)'],
     'Emergence \'21' : targets_2021['Emergence'],
     'Yield \'21' : targets_2021['Combine Yield (t/ac)'],
    }

df = pd.DataFrame(d)

df.plot() # Could change to a scatter plot?
```

```python
# Tilled Data
targets_2020 = tilled_2020[['Emergence', 'Combine Yield (t/ac)']]
targets_2020 = targets_2020.loc[(targets_2020 != 0).all(axis = 1)]
targets_2021 = tilled_2021[['Emergence', 'Combine Yield (t/ac)']]
targets_2021 = targets_2021.loc[(targets_2021 != 0).all(axis = 1)]

d = {'Emergence \'20' : targets_2020['Emergence'],
     'Yield \'20' : targets_2020['Combine Yield (t/ac)'],
     'Emergence \'21' : targets_2021['Emergence'],
     'Yield \'21' : targets_2021['Combine Yield (t/ac)'],
    }

df = pd.DataFrame(d)

df.plot() # Could change to a scatter plot?
```

```python
ploughed_emergence = pd.DataFrame({'Emergence \'20' : ploughed_2020['Emergence'],
    'Emergence \'21' : ploughed_2021['Emergence']
})
tilled_emergence = pd.DataFrame({'Emergence \'20' : tilled_2020['Emergence'],
    'Emergence \'21' : tilled_2021['Emergence']
})

ploughed_yield = pd.DataFrame({'Yield \'20' : ploughed_2020['Combine Yield (t/ac)'],
    'Yield \'21' : ploughed_2021['Combine Yield (t/ac)']
})
tilled_yield = pd.DataFrame({'Yield \'20' : tilled_2020['Combine Yield (t/ac)'],
    'Yield \'21' : tilled_2021['Combine Yield (t/ac)']
})
ploughed_emergence = ploughed_emergence.fillna(0)
tilled_emergence = tilled_emergence.fillna(0)
ploughed_yield = ploughed_yield.fillna(0)
tilled_yield = tilled_yield.fillna(0)


ploughed_emergence_avg = ploughed_emergence[['Emergence \'20', 'Emergence \'21']].mean(axis=1)
tilled_emergence_avg = tilled_emergence[['Emergence \'20', 'Emergence \'21']].mean(axis=1)

ploughed_yield_avg = ploughed_yield[['Yield \'20', 'Yield \'21']].mean(axis=1)
tilled_yield_avg = tilled_yield[['Yield \'20', 'Yield \'21']].mean(axis=1)
```

```python
df = pd.DataFrame({
    'Ploughed Emergence' : ploughed_emergence_avg,
    'Tilled Emergence' : tilled_emergence_avg,
})

df.plot() # Could change to a scatter plot?
```

```python
df = pd.DataFrame({
    'Ploughed Yield' : ploughed_yield_avg,
    'Tilled Yield' : tilled_yield_avg
})

df.plot() # Could change to a scatter plot?
```

Yield noticeably higher for the ploughed plots.
