# Cleaning 2020 Data

```python
df_2020 = df_2020.fillna(0)
```

## Converting Combine Yield From t/hc to t/ac
Multiplying value by 2.471, because 1 hectare is 2.471 acres.

```python
df_2020['Combine Yield (t/ac)'] = df_2020['Combine Yield (t/hc)'].astype(float) * 2.471
del df_2020['Combine Yield (t/hc)'] # Remove hectare col as no longer needed
```

## Cleaning Bulk Density Numbers

```python
bd_moisture_cols = ['BD Moisture 0-10cm', 'BD Moisture 10-20 cm', 'BD Moisture 20-30 cm']

for col in bd_moisture_cols:
    df_2020[col] = df_2020[col].astype(str).map(lambda x: x.rstrip('%')) # Removing %

df_2020
```

# Saving Cleaned Data

```python
df_2020.to_csv('./data/2020.csv')
```
